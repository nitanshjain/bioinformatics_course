```python
with open('/Users/nitanshjain/Documents/Projects/bioinformatics/text.txt') as f:
    genome = f.read().strip()
# print(genome)
text = genome.splitlines()
genome = text[0]
# k = text[1]
# output = text[2]
```

```python
def frequency_mapping_table(genome, k):
    frequency_table = {}

    for i in range(len(genome) - int(k) + 1):
        pattern = genome[i:i+int(k)]
        if pattern not in frequency_table:
            frequency_table[pattern] = 1
        elif pattern in frequency_table:
            frequency_table[pattern] += 1
            
    return frequency_table
            
            
```

```python
n = len(genome)
k = 9
t = 3
l = 500
patterns = []
import json
for i in range(0, n-l):
    window = genome[i:i+l]
    frequency_map = frequency_mapping_table(window, k)
    with open(f'frequency_map_{i}.json', 'w') as f:
        json.dump(frequency_map, f)
    # print("================================")
    # for pattern in frequency_map:
    #     # print(pattern, frequency_map[pattern], t)
    #     if frequency_map[pattern] >= t:
    #         patterns.append(pattern)

print(set(patterns))
    
```
