```python
def hamming_distance(genome1, genome2):
    if len(genome1) != len(genome2):
        raise ValueError("Genomes must be of the same length")
    
    distance = 0
    for base1, base2 in zip(genome1, genome2):
        if base1 != base2:
            distance += 1
            
    return distance
```

```python
def identify_kmer_mismatches(genome1, genome2, k):
    # if len(genome1) != len(genome2):
    #     raise ValueError("Genomes must be of the same length")
    
    mismatches = []
    print(len(genome1), len(genome2))
    for i in range(len(genome2) - len(genome1) + 1):
        kmer1 = genome1
        kmer2 = genome2[i:i+len(genome1)]
        # print(f"Comparing {kmer1} with {kmer2}")
        
        if hamming_distance(kmer1, kmer2) <= int(k):
            mismatches.append((i))
    
    return mismatches
```

```python
def count_pattern_occurence_with_mismatches(genome, pattern, k):
    count = 0
    for i in range(len(genome) - len(pattern) + 1):
        kmer = genome[i:i+len(pattern)]
        if hamming_distance(kmer, pattern) <= int(k):
            count += 1
    return count
```

```python
def immediate_neighbours(pattern, hamming_distance_threshold=1):
    neighborhood = set()
    if hamming_distance_threshold == 0:
        return {pattern}
    
    for i in range(len(pattern)):
        for nucleotide in "ACGT":
            neighbor = pattern[:i] + nucleotide + pattern[i+1:]
            neighborhood.update(immediate_neighbours(neighbor, hamming_distance_threshold - 1))
    
    return neighborhood

def first_symbol(pattern):
    return pattern[0]

def suffix_symbol(pattern):
    return pattern[1:]
```

```python
def frequency_mapping_table_with_mismatches(genome, k, mismatch_threshold):
    frequency_table = {}
    patterns = []

    for i in range(len(genome) - int(k) + 1):
        
        pattern = genome[i:i+int(k)]
        neighbors = immediate_neighbours(pattern, mismatch_threshold)
        
        for neighbor in neighbors:
            
            if neighbor not in frequency_table:
                frequency_table[neighbor] = 1
            else:
                frequency_table[neighbor] += 1
        
    max_frequency = max(frequency_table.values())
    for pattern, frequency in frequency_table.items():
        if frequency == max_frequency:
            patterns.append(pattern)
            
    return frequency_table, patterns

            
            
```

```python
def reverse_complement(dna):
    dna = dna.upper().replace("/n", "")
    complement = {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C'}
    return ''.join(complement.get(base) for base in reversed(dna))
```

```python
def frequency_mapping_table_with_mismatches_complement(genome, k, mismatch_threshold):
    frequency_table = {}
    patterns = []

    for i in range(len(genome) - int(k) + 1):
        
        pattern = genome[i:i+int(k)]
        neighbors = immediate_neighbours(pattern, mismatch_threshold)
        print(f"Pattern: {pattern}")
        
        pattern_complement = reverse_complement(pattern)
        neighbors_complement = immediate_neighbours(pattern_complement, mismatch_threshold)
        
        
        for neighbor in neighbors:
            
            if neighbor not in frequency_table:
                frequency_table[neighbor] = 1
            else:
                frequency_table[neighbor] += 1
                
        for neighbor in neighbors_complement:
            
            if neighbor not in frequency_table:
                frequency_table[neighbor] = 1
            else:
                frequency_table[neighbor] += 1
        
    max_frequency = max(frequency_table.values())
    for pattern, frequency in frequency_table.items():
        if frequency == max_frequency:
            patterns.append(pattern)
            
    return frequency_table, patterns

            
            
```
