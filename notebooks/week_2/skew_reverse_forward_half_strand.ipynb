```python
input_genome = "GAGCCACCGCGATA"
```

```python
def skew_check(genome):
    skew = 0
    skew_list = [0]
    for base in genome:
        if base == 'G':
            skew += 1
        elif base == 'C':
            skew -= 1
        skew_list.append(skew)
    return skew_list
```

```python
output = skew_check(input_genome)
```

```python
output
```

```python
# Create a Graph with the Skew List vs Length of the Genome
from matplotlib import pyplot as plt

plt.plot(output)
plt.xlabel('Position in Genome')
plt.ylabel('Skew (G-C)')
plt.title('Skew Plot of Genome')
plt.grid()
```
