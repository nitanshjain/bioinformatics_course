```python
with open('/Users/nitanshjain/Documents/Projects/bioinformatics/Peculiar Statistics Half-Strands Step 10.txt') as f:
    input_genome = f.read().strip()
# input_genome = "GAGCCACCGCGATA"
```

```python
def skew_check(genome):
    skew = 0
    skew_list = [0]
    for base in genome:
        if base == 'G':
            skew += 1
        elif base == 'C':
            skew -= 1
        skew_list.append(skew)
    return skew_list
```

```python
output_skew = skew_check(input_genome)
```

```python
print(output_skew)
```

```python
# Create a Graph with the Skew List vs Length of the Genome
from matplotlib import pyplot as plt

plt.plot(output_skew)
plt.xlabel('Position in Genome')
plt.ylabel('Skew (G-C)')
plt.title('Skew Plot of Genome')
plt.grid()
```

```python
min_skew_value = min(output_skew)
print("Minimum Skew Value:", min_skew_value)
```

```python
# index of minimum skew value
min_skew_indices = [i for i, x in enumerate(output_skew) if x == min_skew_value]
print("Indices of Minimum Skew Value:", min_skew_indices)
```
