```python
def skew(genome):
    skew = 0
    skew_list = [0]
    for base in genome:
        if base == 'G':
            skew += 1
        elif base == 'C':
            skew -= 1
        skew_list.append(skew)
    return skew_list
```

```python
def return_min_skew_position(skew_list):
    min_skew = min(skew_list)
    min_positions = [i for i, skew in enumerate(skew_list) if skew == min_skew]
    return min_positions
```

```python
def return_max_skew_position(skew_list):
    max_skew = max(skew_list)
    max_positions = [i for i, skew in enumerate(skew_list) if skew==max_skew]
    return max_positions
```

```python
def hamming_distance(genome1, genome2):
    if len(genome1) != len(genome2):
        raise ValueError("Genomes must be of the same length")
    
    distance = 0
    for base1, base2 in zip(genome1, genome2):
        if base1 != base2:
            distance += 1
            
    return distance

print(hamming_distance("CTACAGCAATACGATCATATGCGGATCCGCAGTGGCCGGTAGACACACGT", "CTACCCCGCTGCTCAATGACCGGGACTAAAGAGGCGAAGATTATGGTGTG"))
```

```python
def reverse_complement(dna):
    complement = {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C'}
    return ''.join(complement[base] for base in reversed(dna))
```

```python
def identify_kmer_mismatches(genome1, genome2, mismatch_value):

    mismatches = []
    for i in range(len(genome2) - len(genome1) + 1):
        kmer1 = genome1
        kmer2 = genome2[i:i+len(genome1)]
        
        if hamming_distance(kmer1, kmer2) <= int(mismatch_value):
            mismatches.append((i))
    
    return mismatches
```

```python
def immediate_neighbours(pattern, hamming_distance_threshold):
    
    neighborhood = set()
    if hamming_distance_threshold == 0:
        return {pattern}
    
    for i in range(len(pattern)):
        for nucleotide in "ACGT":
            neighbor = pattern[:i] + nucleotide + pattern[i+1:]
            neighborhood.update(immediate_neighbours(neighbor, hamming_distance_threshold - 1))
    
    return neighborhood
```

```python
def frequency_mapping_table_with_mismatches(genome, k, mismatch_threshold):
    frequency_table = {}
    patterns = []

    for i in range(len(genome) - int(k) + 1):
        
        pattern = genome[i:i+int(k)]
        neighbors = immediate_neighbours(pattern, mismatch_threshold)
        
        for neighbor in neighbors:
            
            if neighbor not in frequency_table:
                frequency_table[neighbor] = 1
            else:
                frequency_table[neighbor] += 1
        
    max_frequency = max(frequency_table.values())
    for pattern, frequency in frequency_table.items():
        if frequency == max_frequency:
            patterns.append(pattern)
            
    return frequency_table, patterns
```

```python
def pick_windows(genome_len: int, anchor: int, L: int, mode: str):
    n = genome_len
    out = []
    if mode == "start":
        s, e = anchor, min(anchor + L, n)
        out.append((s, e))
    elif mode == "end":
        e, s = anchor, max(0, anchor - L)
        out.append((s, e))
    elif mode == "center":
        s = max(0, anchor - L // 2)
        e = min(n, s + L)
        out.append((s, e))
    elif mode == "around":
        out.extend(pick_windows(n, anchor, L, "start"))
        out.extend(pick_windows(n, anchor, L, "end"))
        out.extend(pick_windows(n, anchor, L, "center"))
        out = sorted(set(out))
    else:
        raise ValueError("WINDOW_MODE must be start | end | center | around")
    return out
```

```python
WINDOW = 500
WINDOW_MODE = "around"
K = 9
D = 1
```

```python
from collections import defaultdict
def main():
    genome = final_genome
    genome_skew_list = skew(genome)
    mins = return_min_skew_position(genome_skew_list)
    anchor = mins[0]  # first minimum-skew position
    
    # examine windows and also aggregate across them
    aggregated = defaultdict(int)
    reports = []
    for s, e in pick_windows(len(genome), anchor, WINDOW, WINDOW_MODE):
        sub = genome[s:e]
        freq, best = frequency_mapping_table_with_mismatches(sub, K, D)
        for kmer, c in freq.items():
            aggregated[kmer] += c
        reports.append(((s, e), sorted(best)))

    # global winners across all inspected windows
    winners = []
    if aggregated:
        maxc = max(aggregated.values())
        winners = sorted([p for p, c in aggregated.items() if c == maxc])

    # minimal, compact output
    print("min_skew_positions:", " ".join(map(str, mins[:10])) + (" ..." if len(mins) > 10 else ""))
    print("anchor:", anchor)
    for (s, e), best in reports:
        print(f"window {s}:{e}  best({K},d={D}):", " ".join(best))
    print("global_top:", " ".join(winners))

if __name__ == "__main__":
    main()
```
